````python
import requests
from bs4 import BeautifulSoup

def scrape_wikipedia(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    # Extract the title
    title = soup.find("h1", {"id": "firstHeading"}).text.strip()

    # Extract all paragraphs
    content = ""
    paragraphs = soup.find_all("p")
    for para in paragraphs:
        text = para.text.strip()
        if text:
            content += text + "\n\n"  # Adding spacing for readability

    return title, content

# Example usage
url = "https://en.wikipedia.org/wiki/Muhammad"
title, article_content = scrape_wikipedia(url)

# Save the full article to a .txt file
filename = f"{title.replace(' ', '_')}.txt"
with open(filename, "w", encoding="utf-8") as file:
    file.write(f"Title: {title}\n\n")
    file.write(f"Article Content:\n{article_content}\n")

print(f"Full Wikipedia article saved as {filename}")
````

# Tokenization

````python
# Step1: tokeinzation using re in python
import re

def tokenizer(text):
    """
    Tokenizes text using regex to handle words, contractions, and punctuation.
    """
    pattern = r"\b\w+(?:'\w+)?\b|[.,!?;:()\"]"
    tokens = re.findall(pattern, text)
    tokens = [token.strip() for token in tokens if token.strip()]
    return tokens

def tokenize_file(input_file):
    """
    Reads a text file, tokenizes its content, and returns the tokens as a list.
    """
    with open(input_file, "r", encoding="utf-8") as file:
        text = file.read()

    tokens = tokenizer(text)
    return tokens

# Example usage
input_file = "Muhammad.txt"  # Change to your actual file name
tokens = tokenize_file(input_file)

# Print the tokens
print(len(tokens))
````

````python
# checking total duplicate words
from collections import Counter

# Count word occurrences
word_counts = Counter(tokens)

# Find duplicate words (words with count > 1)
duplicates = {word: count for word, count in word_counts.items() if count > 1}

# Print the total number of duplicate words
print(f"Total duplicate words: {len(duplicates)}")

print("Duplicate words and their counts:")
for word, count in duplicates.items():
  print(f"{word}: {count}")
````

````python
import re

class SimpleTokenizerV1:
    def __init__(self, vocabulary):
        self.str_to_int = vocabulary  # Mapping string to integer
        self.int_to_str = {i: s for s, i in vocabulary.items()}  # Mapping integer to string

        # Ensure "<UNK>" token exists
        if "<UNK>" not in self.str_to_int:
            unk_id = max(self.str_to_int.values(), default=-1) + 1
            self.str_to_int["<UNK>"] = unk_id
            self.int_to_str[unk_id] = "<UNK>"

    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [item.strip() for item in preprocessed if item.strip()]

        # Replace unknown tokens with "<UNK>" ID
        ids = [self.str_to_int.get(s, self.str_to_int["<UNK>"]) for s in preprocessed]

        return ids

    def decode(self, ids):
        # Ensure safe lookup with default "<UNK>"
        text = " ".join([self.int_to_str.get(i, "<UNK>") for i in ids])

        # Remove unnecessary spaces before punctuation
        text = re.sub(r'\s+([,.:;?_!"()\'])', r'\1', text)

        return text
````

- The encode function turns text into token IDs
- The decode function turns token IDs back into text
- We can use the tokenizer to encode (that is, tokenize) texts into integers
- These integers can then be embedded (later) as input of/for the LLM

## 2.4 Adding special context tokens
It's useful to add some "special" tokens for unknown words and to denote the end of a text
image.png

- Some tokenizers use special tokens to help the LLM with additional context

- Some of these special tokens are

=> [BOS] (beginning of sequence) marks the beginning of text\
=> [EOS] (end of sequence) marks where the text ends (this is usually used to concatenate multiple unrelated texts, e.g., two different Wikipedia articles or two different books, and so on)\
=> [PAD] (padding) if we train LLMs with a batch size greater than 1 (we may include multiple texts with different lengths; with the padding token we pad the shorter texts to the longest length so that all texts have an equal length)\
=> [UNK] to represent words that are not included in the vocabulary

Note that GPT-2 does not need any of these tokens mentioned above but only uses an <|endoftext|> token to reduce complexity

The <|endoftext|> is analogous to the [EOS] token mentioned above

GPT also uses the <|endoftext|> for padding (since we typically use a mask when training on batched inputs, we would not attend padded tokens anyways, so it does not matter what these tokens are)

GPT-2 does not use an <UNK> token for out-of-vocabulary words; instead, GPT-2 uses a byte-pair encoding (BPE) tokenizer, which breaks down words into subword units which we will discuss in a later section

1️⃣ tokenizer.special_tokens_set - {"<|endoftext|>"}
tokenizer.special_tokens_set: Contains all special tokens that are typically disallowed in normal encoding/decoding.

Subtracting {"<|endoftext|>"} means removing <|endoftext|> from the disallowed set, so it can be used.

2️⃣ disallowed_special=(...)
This argument tells tiktoken to disallow all special tokens except <|endoftext|>.

As a result, any other special tokens (e.g., <|startoftext|>, <|pad|>, etc.) will still be blocked.

3️⃣ tokenizer.encode(text, disallowed_special=...)
Encodes text, ensuring that <|endoftext|> remains valid while other special tokens are filtered out.

4️⃣ tokenizer.decode(...)
Converts the encoded tokens back into readable text.

````python
# we are using pytorch to load and work with data
# for llarge dataset this is not the correct approach as we donot save trillions of token in memory
from torch.utils.data import Dataset, DataLoader


class GPTDatasetV1(Dataset):
    def __init__(self, txt, tokenizer, max_length, stride):
        self.input_ids = []  # empty token list
        self.target_ids = []

        # Tokenize the entire text
        token_ids = tokenizer.encode(txt, allowed_special={"<|endoftext|>"}) # tokenize the text

        # Use a sliding window to chunk the book into overlapping sequences of max_length
        for i in range(0, len(token_ids) - max_length, stride):  # creating chunks for whole dataset
            input_chunk = token_ids[i:i + max_length]
            target_chunk = token_ids[i + 1: i + max_length + 1]
            self.input_ids.append(torch.tensor(input_chunk))
            self.target_ids.append(torch.tensor(target_chunk))

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        return self.input_ids[idx], self.target_ids[idx]
````

````python
# data loading
def create_dataloader_v1(txt, batch_size=2, max_length=256,
                        stride=128, shuffle=True, drop_last=True,
                        num_workers=0):

    # Initialize the tokenizer
    tokenizer = tiktoken.get_encoding("gpt2")

    # Create dataset
    dataset = GPTDatasetV1(txt, tokenizer, max_length, stride)

    # Create dataloader
    dataloader = DataLoader(
        dataset,
        batch_size=batch_size,
        shuffle=shuffle,
        drop_last=drop_last,
        num_workers=num_workers
    )

    return dataloader

 # batch_size mean how many batch we are providing , max_length mean context_size,
# drop_last mean if after batchsize any data remaining just drop it and num_worker mean no.background processors
````

````python
with open("Muhammad.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()
````

- For those who are familiar with one-hot encoding, the embedding layer approach above is essentially just a more efficient way of implementing one-hot encoding followed by matrix multiplication in a fully-connected layer, which is described in the supplementary code in ./embedding_vs_matmul
- Because the embedding layer is just a more efficient implementation that is equivalent to the one-hot encoding and matrix-multiplication approach it can be seen as a neural network layer that can be optimized via backpropagation
